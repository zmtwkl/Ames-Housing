```python
# import libraries
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.info()
```

# 1. Count of one categorical feature

```python
# preprocessing / Construct a dataset
Neighborhood_count = df.groupby(['Neighborhood'],as_index=False).size()
```

```python
Neighborhood_count
```

```python
sns.barplot(x=Neighborhood_count['Neighborhood'],
            y=Neighborhood_count['size'])
plt.show()
```

```python
# sort values
Neighborhood_count = df.groupby(['Neighborhood'],as_index=False).size()
Neighborhood_count = Neighborhood_count.sort_values(['size'],ascending=False)
```

```python
sns.barplot(x=Neighborhood_count['Neighborhood'],
            y=Neighborhood_count['size'])
plt.show()
```

```python
# rotate text
sns.barplot(x=Neighborhood_count['Neighborhood'],
            y=Neighborhood_count['size'])
plt.xticks(rotation=90)
plt.show()
```

```python
# change axis
sns.barplot(x=Neighborhood_count['size'],
            y=Neighborhood_count['Neighborhood'])
plt.show()
```

# 2. Distribution of one continuous feature

```python
# show histogram and KDE plot
sns.displot(x=df['YearBuilt'],kde=True)
plt.show()
```

# 3. Categorical-continuous

```python
sns.boxplot(x=df['RoofStyle'],
            y=df['OverallCond'])
plt.show()
```

# 4. Continuous-continuous

```python
sns.scatterplot(x=df['OverallQual'],y=df['OverallCond'])
plt.show()
```
